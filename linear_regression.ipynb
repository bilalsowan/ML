```python
# prompt: generate simple example regarding linear regression


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Sample data
X = np.array([[1], [2], [3], [4], [5]])  # Input features
y = np.array([2, 4, 5, 4, 5])  # Target variable

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)


# Print model coefficients
b0 = model.intercept_
b1 = model.coef_[0]

print(f"Intercept (b₀): {b0:.2f}")
print(f"Coefficient (b₁): {b1:.2f}")

# Display the linear regression formula
print(f"Linear Regression Formula: y = {b0:.2f} + {b1:.2f}X")


# Evaluation metrics
# MAE: Mean Absolute Error
# Formula: MAE = (1/n) * Σ|yᵢ - ŷᵢ|
mae = mean_absolute_error(y, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# MSE: Mean Squared Error
# Formula: MSE = (1/n) * Σ(yᵢ - ŷᵢ)²
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# RMSE: Root Mean Squared Error
# Formula: RMSE = √MSE
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# R²: Coefficient of Determination
# Formula: R² = 1 - (Σ(yᵢ - ŷᵢ)² / Σ(yᵢ - ȳ)²)
r2 = r2_score(y, y_pred)
print(f"R-squared (R²): {r2:.2f}")

# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Evaluation Metrics')
plt.legend()
plt.show()

```

```python
# prompt: generate simple example regarding linear regression


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = np.array([[1000], [1500], [2000], [2500], [3000]])  # Area in sq. ft.
y = np.array([200000, 250000, 300000, 350000, 400000])  # Price in USD

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)


# Print model coefficients
b0 = model.intercept_
b1 = model.coef_[0]

print(f"Intercept (b₀): {b0:.2f}")
print(f"Coefficient (b₁): {b1:.2f}")

# Display the linear regression formula
print(f"Linear Regression Formula: y = {b0:.2f} + {b1:.2f}X")


# Evaluation metrics
# MAE: Mean Absolute Error
# Formula: MAE = (1/n) * Σ|yᵢ - ŷᵢ|
mae = mean_absolute_error(y, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# MSE: Mean Squared Error
# Formula: MSE = (1/n) * Σ(yᵢ - ŷᵢ)²
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# RMSE: Root Mean Squared Error
# Formula: RMSE = √MSE
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# R²: Coefficient of Determination
# Formula: R² = 1 - (Σ(yᵢ - ŷᵢ)² / Σ(yᵢ - ȳ)²)
r2 = r2_score(y, y_pred)
print(f"R-squared (R²): {r2:.2f}")

# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Evaluation Metrics')
plt.legend()
plt.show()
```
