```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data
area = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)  # Independent variable (sq. ft.)
price = np.array([200000, 250000, 300000, 350000, 400000])  # Dependent variable (USD)

# Linear Regression Model
model = LinearRegression()
model.fit(area, price)

# Coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Predicted prices
predicted_prices = model.predict(area)

# Plot the data and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(area, price, color='blue', label='Actual Prices')
plt.plot(area, predicted_prices, color='red', label='Regression Line')
plt.xlabel('Area (sq. ft.)')
plt.ylabel('Price (USD)')
plt.title('Linear Regression: Area vs Price')
plt.legend()
plt.grid()
plt.show()

slope, intercept

new_area = np.array([[1800]])  # New area
predicted_price = model.predict(new_area)
print(predicted_price)



# Extracting slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Formatting the formula
formula = f"Price (USD) = {slope:.2f} * Area (sq. ft.) + {intercept:.2f}"
print("Linear Regression Formula:", formula)
```
