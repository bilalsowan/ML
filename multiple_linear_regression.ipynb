```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Step 1: Create the dataset
data = {
    "Area": [1500, 1700, 2000, 2200, 2500],
    "Bedrooms": [3, 4, 3, 5, 4],
    "Age": [20, 15, 10, 5, 1],
    "Price": [300000, 340000, 400000, 450000, 500000]
}
df = pd.DataFrame(data)

# Step 2: Prepare the features (X) and target (Y)
X = df[["Area", "Bedrooms", "Age"]]
Y = df["Price"]

# Step 3: Fit the multiple linear regression model
model = LinearRegression()
model.fit(X, Y)

# Step 4: Get the coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

# Step 5: Print the regression formula
formula = f"Price = {intercept:.2f} + {coefficients[0]:.2f}*Area + {coefficients[1]:.2f}*Bedrooms + {coefficients[2]:.2f}*Age"
print("Regression Formula:", formula)

# Step 6: Make predictions
predictions = model.predict(X)

# Step 7: Calculate performance metrics
mse = mean_squared_error(Y, predictions)
mae = mean_absolute_error(Y, predictions)
rmse = np.sqrt(mse)
r_squared = r2_score(Y, predictions)

print(f"MSE: {mse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R-squared: {r_squared:.2f}")

# Step 8: Plot the actual vs predicted values
plt.figure(figsize=(8, 6))
plt.plot(Y, label="Actual", marker='o', linestyle='--')
plt.plot(predictions, label="Predicted", marker='x', linestyle='-')
plt.title("Actual vs Predicted Prices")
plt.xlabel("Data Points")
plt.ylabel("Price ($)")
plt.legend()
plt.tight_layout()
plt.show()
```
